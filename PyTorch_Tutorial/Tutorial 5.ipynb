# **Dataset and Dataloader Class**

In PyTorch, **Dataset and DataLoader** are core classes used to efficiently handle and load data during model training.

# Dataset Class

The Dataset class represents the entire dataset. It defines how data samples are stored and accessed. A custom dataset must implement two essential methods ->

*   __len__()

    Returns total number of samples in the dataset.

*   __getitem__(index)

    used to fetch particular row.

The Dataset class is responsible only for data storage and retrieval, not batching or shuffling.

# DataLoader Class

The DataLoader class wraps around a Dataset and controls how data is fed to the model.

It provides the following functionalities ->

*   Divide data into mini-batches
*   Shuffles data
*   supports parallel loadin

During training, the DataLoader returns batches of data instead of individual samples.

```python
import torch
from sklearn.datasets import make_classification
```

```python
# Step 1: Create a synthetic classification dataset using sklearn
x, y = make_classification(
    n_samples = 10,       # Number of samples
    n_features = 2,       # Number of features
    n_informative = 2,    # Number of informative features
    n_redundant = 0,      # Number of duplicate features
    n_classes = 2,        # Number of classes
    random_state = 42     # For reproducibility
)
```

```python
# convert numpy to tensors
x_tensor = torch.from_numpy(x)
y_tensor = torch.from_numpy(y)
```

```python
from torch.utils.data import Dataset, DataLoader
```

```python
class Customdataset(Dataset):

  def __init__(self, features, labels):
    self.features = features
    self.labels = labels

  def __len__(self):
    return self.features.shape[0]

  def __getitem__(self, idx):
    return self.features[idx], self.labels[idx]
```

```python
# creating an object of dataset class

dataset = Customdataset(x_tensor, y_tensor)
```

```python
len(dataset)
```

```python
dataset[0]
```

```python
# create an object for dataloader class

dataloader = DataLoader(dataset, batch_size = 2, shuffle = True)
```

```python
# to create batches

for batch_features, batch_label in dataloader:
  print(batch_features)
  print(batch_label)
  print("-"*50)
```
